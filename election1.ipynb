```python
bj=data[data["Party Abbreviation"]=="BJP"]
bjpvote=bj["Total Votes Polled"].sum()
bjpvote
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5))
sns.countplot(data["Candidate Sex"])
plt.show()
```

```python

#barchart  using matplotlib
party=["bjp","inc","ind","bsp","tdp"]
votes=[bjpvote,incpvote,indvote,bspvote,tdpvote]
plt.figure(figsize=(10,5))
plt.title("election data",fontsize=15,color="m")
plt.xlabel("party",fontsize=12,color="b")
plt.ylabel("votes",fontsize=12,color="b")
plt.bar(party,votes,width=0.4,color="cmyk")
```

```python
fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

party=["bjp","inc","ind","bsp","tdp"]

votes=[bjppercentage,incpercentage,indpercentage,bsppercentage,tdppercentage]

wedges, texts = ax.pie(votes, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(party[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Matplotlib election: A donut")

plt.show()
```
